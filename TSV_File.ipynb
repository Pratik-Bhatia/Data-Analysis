TSV With The Simpsons episodes
(Tabular Separated Values)

```python
import pandas as pd
import numpy as np

col_names = ['Title', 'Air date', 'Production code', 'Season', 'Number in season',
             'Number in series', 'US viewers (million)', 'Views', 'IMDB rating']

select_cols=['Title','Air date','Production code','IMDB rating']

df=pd.read_csv('/content/simpsons-episodes.tsv',
               sep='\t',
               names=col_names,
              usecols=select_cols,
               skiprows=4,
               index_col='Production code',
               na_values=['no_val'],
               )
df
```
